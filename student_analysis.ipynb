The Matplotlib is imported for data visualization.

```python
import matplotlib.pyplot as plt
```

Histogram of final grades.

```python
plt.figure()
plt.hist(a["G3"],bins=15)
plt.xlabel("Final Grade (G3)")
plt.ylabel("Number of Students")
plt.title("Histogram of Final Grade")
plt.show()
```

Study Time VS Final Grades

```python
plt.figure()
plt.scatter(a["studytime"],a["G3"])
plt.xlabel("Study Time")
plt.ylabel("Final Grade(G3)")
plt.title("Study Time VS Final Grade")
plt.show()
```

Avg. Final grade by gender.

```python
gender_avg =a.groupby("sex")["G3"].mean()
plt.figure()
gender_avg.plot(kind="bar")
plt.xlabel("Gender")
plt.ylabel("Avg Final Grade(G3)")
plt.title("Avg Final Grade by Gender")
plt.show()
```
